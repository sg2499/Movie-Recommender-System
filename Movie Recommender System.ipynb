## Data Preprocessing

```python
# Checking for missing data
df.isnull().sum()
```

**NOTE** - Now in our case 'overview' is a very important column, so we will go ahead and drop the rows with missing values.

```python
# Dropping the records with missing values
df = df.dropna()
df.isnull().sum()
```

```python
# Checking for duplicate values if any
df.duplicated().sum()
```

**NOTE** - So as we can see above there are no duplicate values in our dataset.

#### Now we go ahead and preprocess the 'genres' and 'keywords' column.

```python
# Creating a helper function to achieve this task
from ast import literal_eval
def convert(obj):
    genre_list = []
    for i in literal_eval(obj):
        genre_list.append(i['name'])
    return genre_list
```

```python
# Applying our helper function to the 'genres' column in our dataset
df['genres'] = df['genres'].apply(convert)
df.head(3)
```

```python
# Applying the helper function to the 'keywords' column in our dataset
df['keywords'] = df['keywords'].apply(convert)
df.head(3)
```

### Extracting the top 3 cast members from the 'cast' column.

```python
# Creating a helper function to achieve the task
def top_3_cast(obj):
    cast_list = []
    counter = 0
    for i in literal_eval(obj):
        if counter != 3:
            cast_list.append(i['name'])
            counter += 1
        else:
            break
    return cast_list
```

```python
# Extracting the top 3 cast members using our helper function
df['cast'] = df['cast'].apply(top_3_cast)
df.head()
```

### Extracting the name of the director of each movie from the 'crew' column.

```python
# Creating a helper function to achieve this task
def fetch_director(obj):
    director_list = []
    for i in literal_eval(obj):
        if i['job'] == 'Director':
            director_list.append(i['name'])
            break
    return director_list
```

```python
# Applying the helper function to our column 'crew' in the dataset
df['crew'] = df['crew'].apply(fetch_director)
df.head()
```

### Finally we convert the 'overview' column into a list as well.

```python
# Converting the 'overview' column
df['overview'] = df['overview'].apply(lambda x : x.split())
df.head()
```

### Now we move on and fix the newly preprocessed columns and remove spaces if any between two words in order to make them a single entity so that our Recommender System does not get confused after it's deployment.

```python
# Applying the same to our columns
df['genres'] = df['genres'].apply(lambda x : [i.replace(' ','') for i in x])
df['keywords'] = df['keywords'].apply(lambda x : [i.replace(' ','') for i in x])
df['cast'] = df['cast'].apply(lambda x : [i.replace(' ','') for i in x])
df['crew'] = df['crew'].apply(lambda x : [i.replace(' ','') for i in x])
df.head()
```

### Finally, we go on and create our 'tags' column which will be the combination of 'overview', 'genres', 'keywords', 'cast' and 'crew' columns.

```python
# Creating the 'tags' column
df['tags'] = (df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew'])
df.head()
```

```python
# Then we select the required columns to create our final database which would be used in creating our Recommender System
final_df = df[['movie_id', 'title', 'tags']]
final_df.head()
```

```python
# Now we convert the 'tags' column from a list to a string
final_df['tags'] = final_df['tags'].apply(lambda x : ' '.join(x))
final_df.head()
```

```python
# Finally we convert everything present in our 'tags' column to lower case
final_df['tags'] = final_df['tags'].apply(lambda x : x.lower())
final_df.head()
```

```python
final_df['tags'][99]
```

### Now we will apply stemming method to our 'tags' column using the Porter Stemmer.

```python
# Importing the Porter Stemmer and intiating it's object
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

# Now we create our helper function to achieve the task
def stem(text):
    stem_list = []
    for i in text.split():
        stem_list.append(ps.stem(i))
    return ' '.join(stem_list)
```

```python
# Now we go ahead and apply our helper function to our 'tags' column
final_df['tags'] = final_df['tags'].apply(stem)
final_df.head()
```

## Text Vectorization

In this part we will convert our 'tags' column into vectors in order to make the recommender system understand which movies are similar to each other and which are not.

```python
# Creating the vector and applying it to the 'tags' column
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
vectors = cv.fit_transform(final_df['tags']).toarray()
```

```python
# Checking the shape
vectors.shape
```

```python
# Checking out the top 5000 words selected by our vectorizer
print(cv.get_feature_names_out())
```

### Now we move on to calculate the similarities between each movie by calculating their cosine similarities.

```python
# Calculating the cosine similarities
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

### Now we move on to making our main recommender function which will recommend the user top 5 similar movies on providing a movie of his/her choice.

```python
# Creating the function
def recommend_movies(movie):
    movie_index = final_df[final_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x : x[1])[1:6]

    for i in movies_list:
        print(final_df.iloc[i[0]].title)
```
